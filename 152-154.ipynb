```python
# eksik değer var mı?
df.isnull().values.any() 
```

```python
# hangi değişkende kaçar tane eksik değer var?
df.isnull().sum()
```

```python
# eksik değerler hesaplamasa problem yaratabilir. bu yüzden bir değer atanması gerekebilir.
# df['sütun_ismi'].fillna(değer, inplace = True) => belirlediğimiz sütundaki eksik hücrelere bizim atadığımız değerleri yazıyor
```

```python
df["orbital_period"].fillna(0, inplace = True)
df.isnull().sum() # o sütunda artık eksik değer yok
```

```python
# tabiiki bu değer rastgele olamaz. o boşlukları anlamlı bir şekilde doldurmamız gerekir.
df["mass"].fillna(df["mass"].mean(), inplace = True)
df.isnull().sum()
```
